```python
import pandas as pd
```

```python
data = pd.read_csv('../data/spam.csv', encoding='latin-1')
```

```python
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of spam (1) and ham (0)
data['label'] = data['label'].map({'ham': 0, 'spam': 1})  # Convert labels to numeric
sns.countplot(x=data['label'], palette="coolwarm")

# Add labels
plt.title("Spam vs Ham Distribution")
plt.xlabel("Label (0: Ham, 1: Spam)")
plt.ylabel("Count")
plt.show()
```

```python
from wordcloud import WordCloud

# Filter spam messages
spam_messages = " ".join(data[data['label'] == 1]['message'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color="black").generate(spam_messages)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Spam Emails")
plt.show()
```
